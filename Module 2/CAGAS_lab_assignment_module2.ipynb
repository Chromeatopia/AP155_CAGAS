**Calculate the inverse of a matrix and LU decomposition**

1. Use numpy's $\tt linalg$ package to calculate the inverse of the ff. matrix *(5 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
4 & -2 & 1 \\
3 & 6 & -4 \\
2 & 1 & 8 
\end{pmatrix} \end{equation} $

2. Check your inverse in both directions; that is, check that ${\bf AA}^{−1} = {\bf A}^{−1}{\bf A} = {\bf I}$. Up to how many number of decimal places is this true? *(5 pts.)*

3. Let this matrix $\bf A$ be used to describe three simultaneous linear equations, ${\bf Ax = b}$. Solve for the vector $\bf x$ given
${\bf b} = \begin{pmatrix} 
12 \\ 
-25 \\ 
32 
\end{pmatrix}$ using: (i) the numerical inverse calculated in (1) and (ii) LU decomposition and backsubstitution with numpy's $\tt linalg$ package. *(5 pts. each)*


```python

# Import required package
import numpy as np
import numpy.linalg as la
from numpy.linalg import solve


A = np.array([[4, -2, 1], 
              [3, 6, -4],
              [2, 1, 8]])

print("/*Number 1*/")
# Make A inverse using linalg
Ainv = np.linalg.inv(A)

print("The inverse of Matrix A is: ")
print(Ainv)
print(" ")

print("/*Number 2*/")
# Check inverse using A * A-1 = I
I1 = np.matmul(A, Ainv)

print("Checking the Identity Matrix using A * A-1 we get: ")
print(I1)
print(" ")

# Check inverse using A-1 * A = I
I2 = np.matmul(Ainv, A)

print("Checking the Identity Matrix using A-1 * A we get: ")
print(I2)
print(" ")

print("The answer is true up to at least 16 decimal places!")
print(" ")

print("/*Number 3.1*/")
# Solve x at A * x = b using A inverse

b = np.array ([[12],
              [-25],
              [32]])

x1 = np.matmul(Ainv, b)

print("Using the inverse of A, x is equal: ")
print(x1)
print(" ")

print("/*Number 3.2*/")
# Solve x at A * x = b using LU decomposition and backsubstitution

x = solve(A, b)

print("Using the linalg.solve function, x is equal: ")
print(x)
print(" ")

# LU decomposition 
```

Assuming circular orbits, and assuming that the Earth is much more massive than either the Moon or the satellite, the distance $r$ from the center of the Earth to the $L_1$ point satisfies
$$ \begin{equation}
\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r,
\end{equation} $$
where $R$ is the distance from the Earth to the Moon, $M$ and $m$ are the Earth and Moon masses, $G$ is Newton's gravitational constant, and $\omega$ is the angular velocity of both the Moon and the satellite.

Use either one of (i) Newton's method or (ii) the secant method to solve for the distance r from the Earth to the $L_1$ point. Compute a solution accurate to at least four significant figures. (30 pts.) 

```python
# Define Secant Function

def secantmethod(func, x0, x1, n):
    
    def f(x): # This is for a more generalized code
        f = eval(func) # It makes it so that whatever function is placed under 'func', it becomes the function being evaluated
        return f
    
    for intercept in range(1,n):
        fx0 = f(x0)
        fx1 = f(x1)
        
        xi = x0 - (fx0 / ((fx0-fx1) / (x0-x1))) # Secant Method formula
        
        x0 = x1
        x1 = xi
        
    print(f"The root was found to be at {xi} after {n} iterations!")
```
