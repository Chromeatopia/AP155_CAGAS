### PROBLEM 1 (20 pts.; 50/50 for code & discussion) 
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(10 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(10 pts.)*


````python
# Set up imports things
from numpy import sin, sqrt
import math
from numpy import zeros
from cmath import exp, pi
import matplotlib.pyplot as plt
import numpy as np

# set up the function

def f(n):
  return (sin((pi)*n)/N)*(sin(20*(pi)*n)/N)

N = 1000
n = np.linspace(0, 1, 1000)

plt.plot (n, f(n))

````

````python
from numpy import loadtxt
from pylab import plot, xlim, show

def dft(y):
    N = len(y)
    c = zeros(N//2+1, complex)
    for k in range (N//2+1):
        for n in range (N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c

y = (sin((pi)*n)/N)*(sin(20*(pi)*n)/N)
c = dft(y)
plot(abs(c), 'o--')
xlim(0, 40)
show()
````

````python
from numpy import loadtxt
import numpy as np
import matplotlib.pyplot as plt
from cmath import exp, pi

blur = np.loadtxt("blur.txt")

plt.imshow(blur, cmap="gray")
````

````python
# Gaussian Filter

def gaussian_filter (k, sigma):
    arx = np.arange((-k//2) + 1.0, (k//2) + 1.0)
    x, y = np.meshgrid(arx, arx)
    return (np.exp(-((np.square(x) + np.square(y))/(2*np.square(sigma)))))
            
````

````python
g1 = gaussian_filter (k=1024, sigma = 60)
plt.imshow(g1, cmap='gray')
````

````python
g2 = np.fft.fftshift(g1)
plt.imshow(g2, cmap='gray')
````

````python
blr = np.fft.rfft2(blur)
gss = np.fft.rfft2(g2)

np.seterr(divide='ignore', invalid='ignore') 

# Getting rid of zeroes
# gss[256, 0] = gss[255, 0]
# gss[512, 0] = gss[513, 0]
# gss[768, 0] = gss[769, 0]


div = np.divide(blr, gss)

deb = np.fft.irfft2(div)

plt.imshow (deb, cmap='gray')
````
